```python
# Dependencies and Setup
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"


# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drugdata = pd.read_csv(mouse_drug_data_to_load)
clintrialdata = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combinedata=pd.merge(mouse_drugdata,clintrialdata,on ='Mouse ID',how ='outer')

# Display the data table for preview
combinedata.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
meantumorvalue=combinedata.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean().reset_index()
meantumorvalue=meantumorvalue.rename(columns={"Tumor Volume (mm3)": "Mean Tumor Volume (mm3)"})
meantumorvalue.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame
stderrortumorvalue=combinedata.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem().reset_index()
stderrortumorvalue=stderrortumorvalue.rename(columns={"Tumor Volume (mm3)":"Standard Error of Tumor Volumes"})
stderrortumorvalue.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#check meantumorvalue shape and dtypes to ensure numbers are numbers(floats/integers) and there are no missing values
meantumorvalue.shape
meantumorvalue.dtypes
#ensure 10 timepoints from 5 minutes to 45 minutes are available for all 10 drugs
meantumorvalue["Timepoint"].value_counts()
#use the pivot feature to set index to "timepoint" and column headers to the 10 drugs included in the trial
rearranged=pd.pivot_table(meantumorvalue, index= ["Timepoint"],columns= ["Drug"])
#remove the 'mean tumor volume' from multiindex column head
rearranged.columns = rearranged.columns.droplevel(0)
#preview
rearranged.head()


```

```python
#use the pivot feature as used above for the std error table
rearrangedsem=pd.pivot_table(stderrortumorvalue, index= ["Timepoint"],columns= ["Drug"])
#remove the 'Standard Error of tumor volumes' from multiindex column head
rearrangedsem.columns = rearrangedsem.columns.droplevel(0)
# Preview that Reformatting worked
rearrangedsem.head()

# Generate the Plot (with Error Bars)
#create a new df for both mean and sem with just drugs/placebo of interest 
reduced = rearranged[["Capomulin","Infubinol","Ketapril","Placebo"]]
reducedsem=rearrangedsem[["Capomulin","Infubinol","Ketapril","Placebo"]]

#scatter plot generation
plt.errorbar(reduced.index,reduced["Capomulin"],yerr=reducedsem["Capomulin"],marker='o',color="r")
plt.errorbar(reduced.index,reduced["Infubinol"],yerr=reducedsem["Infubinol"],marker='^',color="b")
plt.errorbar(reduced.index,reduced["Ketapril"],yerr=reducedsem["Ketapril"],marker='s',color="g")
plt.errorbar(reduced.index,reduced["Placebo"],yerr=reducedsem["Placebo"],marker='d',color="k")
plt.grid()
plt.legend(loc="best")
plt.title("Tumor Response To Treatment")
plt.ylabel("Mean Tumor Volume(mm3)")
plt.xlabel("Timepoint(in days)")
# Save the Figure
plt.savefig("Tumor Response to Treatment.jpeg")
# Show the Figure
plt.show()
# Save the Figure

```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
metaresponsevalue=combinedata.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean().reset_index()
metaresponsevalue=metaresponsevalue.rename(columns={"Metastatic Sites": "Mean of Metastatic Sites"})
metaresponsevalue.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint. 
# Convert to DataFrame
# Preview DataFrame
# not having reset_index will result in a output similar to the output shown a step down
metaerror=combinedata.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem().reset_index()
metaerror=metaerror.rename(columns={"Metastatic Sites": "Std Error, Metastatic Sites"})
metaerror.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
#use the pivot feature as used above for metaresponsevalue table/df
rearrangedmeta=pd.pivot_table(metaresponsevalue, index= ["Timepoint"],columns= ["Drug"])
#remove the 'Mean of metastatic sites' from multiindex column head
rearrangedmeta.columns = rearrangedmeta.columns.droplevel(0)
# Preview that Reformatting worked
rearrangedmeta.head()
#use the pivot feature to set index to "timepoint" and column headers to the 10 drugs included in the trial
rearrangedmetaerror=pd.pivot_table(metaerror, index= ["Timepoint"],columns= ["Drug"])
#remove the 'std error, metastatic sites' from multiindex column head
rearrangedmetaerror.columns = rearrangedmetaerror.columns.droplevel(0)
#preview
rearrangedmetaerror.head()
```

```python
#this output is the same as that of 'rearranged' df
```

```python
#create a new df for both mean and sem of metastatic sites with just drugs/placebo of interest 
reducedmeta = rearrangedmeta[["Capomulin","Infubinol","Ketapril","Placebo"]]
reducedmetasem=rearrangedmetaerror[["Capomulin","Infubinol","Ketapril","Placebo"]]


# Generate the Plot (with Error Bars)
plt.errorbar(reducedmeta.index,reducedmeta["Capomulin"],yerr=reducedmetasem["Capomulin"],marker='o',color="r",linewidth=0.5)
plt.errorbar(reducedmeta.index,reducedmeta["Infubinol"],yerr=reducedmetasem["Infubinol"],marker='^',color="b",linewidth=0.5)
plt.errorbar(reducedmeta.index,reducedmeta["Ketapril"],yerr=reducedmetasem["Ketapril"],marker='s',color="g",linewidth=0.5)
plt.errorbar(reducedmeta.index,reducedmeta["Placebo"],yerr=reducedmetasem["Placebo"],marker='d',color="k",linewidth=0.5)
plt.grid()
plt.legend(loc="best")
plt.title("Metastatic Spread During Treatment")
plt.ylabel("Metastatic Sites",fontsize=9)
plt.xlabel("Timepoint(in days)")
# Save the Figure
plt.savefig("Metastatic Spread During Treatment.jpeg")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame
#nunique () only counts the 'unique' mouse IDs and drops the duplicates
survivalrate=combinedata.groupby(["Drug","Timepoint"])["Mouse ID"].nunique().reset_index()
survivalrate=survivalrate.rename(columns={"Mouse ID": "Mouse Count"})
survivalrate.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#use the pivot feature as used above for metaresponsevalue table/df
rearrangedmousecount=pd.pivot_table(survivalrate, index= ["Timepoint"],columns= ["Drug"])
# Preview the Data Frame
rearrangedmousecount.head()
#remove the 'Mouse Count' from multiindex column head
rearrangedmousecount.columns = rearrangedmousecount.columns.droplevel(0)
# Preview the Data Frame
rearrangedmousecount.head()
```

```python
#reduced df with just the drugs of interest
reducedcount=rearrangedmousecount[["Capomulin","Infubinol","Ketapril","Placebo"]]

# Generate the Plot (Accounting for percentages)
# reducedcount["Capomulin"]/reducedcount["Capomulin"][0])*100 portion for each drug calculates the % value needed for plotting
plt.errorbar(reducedcount.index,(reducedcount["Capomulin"]/reducedcount["Capomulin"][0])*100,yerr=reducedmetasem["Capomulin"],marker='o',color="r",linewidth=0.5)
plt.errorbar(reducedcount.index,(reducedcount["Infubinol"]/reducedcount["Infubinol"][0])*100,yerr=reducedmetasem["Infubinol"],marker='^',color="b",linewidth=0.5)
plt.errorbar(reducedmeta.index,(reducedcount["Ketapril"]/reducedcount["Ketapril"][0])*100,yerr=reducedmetasem["Ketapril"],marker='s',color="g",linewidth=0.5)
plt.errorbar(reducedmeta.index,(reducedcount["Placebo"]/reducedcount["Placebo"][0])*100,yerr=reducedmetasem["Placebo"],marker='d',color="k",linewidth=0.5)
plt.grid()
plt.legend(loc="best")
plt.title("Survival Rate During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Timepoint(in days)")
# Save the Figure
plt.savefig("Survival Rate During Treatment.jpeg")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
#rearranged df was created above where tumor volume data by drug is stored in an ascending order in a drug-specific column
#iloc feature can thus be used to find the last(ending) and the first(starting) tumor volumne to calculate the % value
percentchange = ((rearranged.iloc[-1]-rearranged.iloc[0])/rearranged.iloc[0])*100
# Drop a lisplay the data to confirm
percentchange
```

```python
# Store all Relevant Percent Changes into a Tuple
drugsofinterest=(percentchange["Capomulin"],percentchange["Infubinol"],percentchange["Ketapril"],percentchange["Placebo"])

#drug list for tick names
druglist=["Capomulin","Infubinol","Ketapril","Placebo"]

# Orient widths. Add labels, tick marks, etc. 
ax=plt.axes()
summarygraph=ax.bar(druglist,drugsofinterest,align='edge',color=['g','r','r','r'],width=1)
plt.ylim(-30,70)
ax.grid(which='major')
plt.title("Tumor Change over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
x=np.arange(1,5,1)
ax.set_xticks(x)
ax.set_xticklabels(druglist)

#define/tweak function from source mentioned
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            plt.text(rect.get_x() + rect.get_width()/2., 0.1*height,
                '%d' % int(height) + '%',
                ha='center', va='bottom',color='white')
        else:
            plt.text(rect.get_x() + rect.get_width()/2., 0.5*height,
                '%d' % int(height) + '%',
                ha='center', va='bottom',color='white')
#call the function
autolabel(summarygraph)

#save the figure
plt.savefig("Tumor Change Over 45 Day Treatment.jpeg")

# Show the Figure
plt.show()
```
